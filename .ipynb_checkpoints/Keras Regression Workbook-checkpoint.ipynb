Feature Engineering and getting rid of unimportant features

```python
df = df.drop('id', axis=1)
```

```python
df.head()
```

```python
# convert date from string to an actual datetime object.
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda x: x.year)
df['month'] = df['date'].apply(lambda x: x.month)
```

```python
df.head()
```

```python
# check whether the month and year have a huge impact on the price of the house.
```

```python
# year
plt.figure(figsize=(8, 5))

df.groupby('year').mean()['price'].plot()
```

```python
# month
plt.figure(figsize=(8, 5), dpi=120)

df.groupby('month').mean()['price'].plot()
```

```python
df = df.drop('date', axis=1)
```

```python
df = df.drop('zipcode', axis=1)
```

```python
df['yr_renovated'].value_counts()
```

```python
df.columns
```

Data Preprocessing and Model Creation

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
# test set not fitted to avoid data leakage. It is only transformed.
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df.info()
```

```python
# create model
model = Sequential()

# add layers to model
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

# add final layer for the output of the input and hidden layers
model.add(Dense(1))

# configure model for training
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, 
          batch_size=128, validation_data=(X_test, y_test), 
          epochs=400)
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.head()
```

```python
loss.plot(figsize=(8, 5))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predictions))
```

```python
rmse
```

```python
df['price'].describe()
```

```python
# find out how much variance is being explained by the model.
explained_variance_score(y_test, predictions)
```

```python
plt.figure(figsize=(13, 6), dpi=100)

plt.scatter(y_test, predictions)
plt.xlabel('y_test values')
plt.ylabel('predictions values')

# perfect fitting line
plt.plot(y_test, y_test, c='r')
```
